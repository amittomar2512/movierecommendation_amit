```python
import pandas as pd
import numpy as np
```

```python
movie=pd.read_csv('tmdb_5000_movies.csv')
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head(1)
```

```python
credit.head(1)
```

```python
movies=pd.merge(movie,credit,on='title')
```

```python
pd.set_option('display.max_columns',50)
movies.head(1)
```

```python
moviesR=movies.drop(['budget','homepage','original_language','original_title','popularity','production_companies','production_countries','release_date','revenue','runtime','spoken_languages','status','tagline','vote_average','vote_count','movie_id'],axis=1)
```

```python
moviesR.head(1)
```

```python
import ast 
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
moviesR['genres']=moviesR['genres'].apply(convert)
```

```python
moviesR['keywords']=moviesR['keywords'].apply(convert)
```

```python
moviesR.head(1)
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
moviesR['cast']=moviesR['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
moviesR['crew']=moviesR['crew'].apply(fetch_director)
```

```python
moviesR['overview'][0]
```

```python
moviesR['overview']=moviesR['overview'].apply(lambda x:str(x).split())
```

```python
moviesR.head(1)
```

```python
moviesR['genres']=moviesR['genres'].apply(lambda x:[i.replace(" ","") for i in x])                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 
```

```python
moviesR['keywords']=moviesR['keywords'].apply(lambda x:[i.replace(" ","") for i in x]) 
```

```python
moviesR['overview']=moviesR['overview'].apply(lambda x:[i.replace(" ","") for i in x]) 
```

```python
moviesR['cast']=moviesR['cast'].apply(lambda x:[i.replace(" ","") for i in x]) 
```

```python
moviesR['Tag']=moviesR['genres']+moviesR['keywords']+moviesR['cast']+moviesR['crew']
```

```python
moviesR.head(1)
```

```python
new_movies=moviesR[['id','title','Tag']]
```

```python
new_movies.head(1)
```

```python
new_movies['Tag']=new_movies['Tag'].apply(lambda x:" ".join(x))
```

```python
new_movies.head()
```

```python
new_movies['Tag']=new_movies['Tag'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_movies['Tag']=new_movies['Tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['Tag']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_movies[new_movies['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
